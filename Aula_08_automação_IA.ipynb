```python
# Define uma função para verificar a temperatura do sistema
def check_temperature(temperature):
    # Verifica se a temperatura é maior que 30 graus Celsius
    if temperature > 30:
        # Exibe um alerta de superaquecimento se a condição for verdadeira
        print("ALARME: Superaquecimento detectado! Temperatura acima de 30°C.")
    else:
        # Se a temperatura não exceder 30°C, exibe a temperatura atual e indica que está normal
        print(f"Temperatura atual: {temperature}°C. Tudo normal.")

# --- Exemplo de uso da função com entrada do usuário ---

# Solicita ao usuário que insira a temperatura
try:
    user_temperature = int(input("Por favor, insira a temperatura atual do sistema: "))
    # Chama a função check_temperature com a temperatura fornecida pelo usuário
    check_temperature(user_temperature)
except ValueError:
    print("Entrada inválida. Por favor, insira um número inteiro para a temperatura.")
```
