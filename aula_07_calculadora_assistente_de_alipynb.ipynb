# **calculadora simples**

```python
def soma(num1, num2):
    return num1 + num2

def subtrai(num1, num2):
    return num1 - num2
```

```python
# Define a função para somar dois números
def soma(num1, num2):
    return num1 + num2

# Define a função para subtrair dois números
def subtrai(num1, num2):
    return num1 - num2

# Solicita ao usuário o primeiro número e o converte para float
num1 = float(input("Digite o primeiro número: "))
# Solicita ao usuário o segundo número e o converte para float
num2 = float(input("Digite o segundo número: "))

# Solicita ao usuário a operação desejada (somar ou subtrair) e converte para minúsculas
opcao = input("Deseja 'somar' ou 'subtrair'? ").lower()

# Verifica a opção escolhida pelo usuário e executa a operação correspondente
if opcao == 'somar':
    # Chama a função soma e armazena o resultado
    resultado = soma(num1, num2)
    # Imprime o resultado da soma
    print(f"A soma é: {resultado}")
elif opcao == 'subtrair':
    # Chama a função subtrai e armazena o resultado
    resultado = subtrai(num1, num2)
    # Imprime o resultado da subtração
    print(f"A subtração é: {resultado}")
else:
    # Se a opção for inválida, imprime uma mensagem de erro
    print("Opção inválida. Por favor, escolha 'somar' ou 'subtrair'.")
```
